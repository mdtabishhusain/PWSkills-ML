Q3. How does increasing the value of epsilon affect the number of support vectors in SVR?


In regression problem there are continuous values instead of categories. Therefore the value of marginal plane also changes. We know the equation for marginal plane is:
w^T x+b= +1               for upper values
w^T x+b=-1                for lower values
But since there are continuous values in regression so we use a new equation:
marginal plane=w^T x+b+ε              for upper values
marginal plane=w^T x+b-ε              for lower values
Here,
ε is the marginal error which is a hyperparameter and it lies between best fit line and respective marginal plane
Note that the best fit line is denoted by the equation w^T x+b as shown in the graph. The main aim here is to get maximum distance between the marginal planes.

Here we will use the same cost function like we did in SVC but the constraints will change. So the equation will be 
Cost function=minimize⁡(w.b)  ‖w‖/2+C_i ∑_(i=1)^n▒ζ_i 
The equation C_i ∑_(i=1)^n▒ζ_i  is called hinge loss. But here the meaning of C and ζ will change
The constraint (condition) here is that the difference between the real data point and predicted data point should be less than or equal to marginal error, i.e. 
|y_i-w^T x_i |≤ε
But if we look closely in the graph, there are values which are outside the range of marginal which makes our constraint vulnerable. To tackle this, we will consider these data points as hyperparameter. So our new equation will be:
|y_i-w^T x_i |≤ε+ζ_i
Here,
ζ_i is the error above the margin which is distance between the data points outside marginal plane and the respective marginal plane.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

### Hyperparameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1, 1, 10, 100, 1000],
           'gamma':[1, 0.1, 0.01, 0.001, 0.0001],}
```

```python
grid = GridSearchCV(SVC(),param_grid=param_grid, refit=True, cv= 5, verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4=grid.predict(X_test)
```

```python
y_pred4
```

```python
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
print(accuracy_score(y_test,y_pred4))
```
